```python
import os
root_dir = '/Users/bytedance/workspace/dama/images'
output_path = '/Users/bytedance/workspace/dama/output'
black_list = ['姓名', '性别', '出生', '地址', '婚姻', '电话', '电谌', '年龄', '岁', '身份']
need_extend = ('姓名', '性别', '年龄', '身份证号')

for file in os.listdir(root_dir):
    file_path = root_dir + "/" + file
    if os.path.isdir(file_path) and "DS_Store" not in file:
        for riqi_file in os.listdir(file_path):
            riqi_file_path = file_path + "/" + riqi_file
            if os.path.isdir(riqi_file_path) and "DS_Store" not in riqi_file:
                for tif_image_file in os.listdir(riqi_file_path):
                    if "Thumbs" in tif_image_file:
                        continue
                    jpg_file_path = output_path + "/" + tif_image_file[:-3] + "jpg"
                    jpg_text_path = output_path + "/" + tif_image_file[:-3] + "txt"
                    jpg_text_file = open(jpg_text_path, 'r')
                    result = []
                    for line in jpg_text_file.read_lines():
                        result.append(line.strip().split(','))
                    
                    
                    img = Image.open(jpg_file_path)
                    draw_img = ImageDraw.Draw(img)
                    for res in result:
                        need_mark = False
                        for black in black_list:
                            if black in res[0]:
                                need_mark = True
                                break
                        x1, y1 = int(res[1]), int(res[2]) 
                        x2, y2 = int(res[3]), int(res[4])
                        if res[0].strip() in need_extend:
                            x2 = x1 + (x2 - x1) * 3
                        if need_mark:
                            draw_img.rectangle([(x1, y1), (x2, y2)], fill ="#ff3333", outline ="red")


                    origin_img = Image.open(jpg_file_path)
                    origin_img_size = origin_img.size
                    final_img = Image.new('RGB', (origin_img_size[0] * 2, origin_img_size[1]))
                    final_img.paste(origin_img, (0, 0, origin_img_size[0], origin_img_size[1]))
                    final_img.paste(img, (origin_img_size[0], 0, origin_img_size[0]*2, origin_img_size[1]))
                    final_img.save(jpg_file_path[:-4] + "_final.jpg")
```
